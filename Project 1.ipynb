```python
import pandas as pd
```

```python
file_name = "project 1.csv"
data = pd.read_csv(file_name)
print(data)
```

```python
data['Total Flights'] = data.sum(axis=1)
```

```python
data['% Delayed'] = data['Delayed'].div(data['Total Flights'], axis=0) * 100
print(data)
```

```python
data = data.sort_values(by=['% Delayed'])
data
```
